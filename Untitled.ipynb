```python
def playlist_tracks(username, playlist_id):
    playlist_tracks = spotifyObject.user_playlist_tracks(user=username, playlist_id=playlist_id)
    tracks = playlist_tracks['items']
    while spotifyObject.next(playlist_tracks):
        playlist_tracks = spotifyObject.next(playlist_tracks)
        tracks.extend(playlist_tracks['items'])
    return tracks
```

```python
playlist_dict = {}

for val in range(len(playlists['items'])):
    playlist_dict[playlists['items'][val]['name']] = playlist_tracks(username, playlists['items'][val]['id'])
```

```python
def playlist_tracks_df(playlist_d, key):
    df = pd.DataFrame()
    for idx in range(len(playlist_d[key])):
        df.loc[idx, 'Playlist'] = key
        df.loc[idx, 'Added Date'] = playlist_d[key][idx]['added_at']
        df.loc[idx, 'Artist'] = playlist_d[key][idx]['track']['artists'][0]['name']
        df.loc[idx, 'Id'] = playlist_d[key][idx]['track']['id']
        df.loc[idx, 'Name'] = playlist_d[key][idx]['track']['name']
        df.loc[idx, 'Album'] = playlist_d[key][idx]['track']['album']['name']
    return df
        
```
