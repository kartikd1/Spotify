```python
def get_week_endpoints(df):
    days = df['endTime'].weekday()
    week_start = df['endTime'] - timedelta(days)
    week_end = week_start + timedelta(days=6)
    df['Week'] = str(week_start.strftime('%Y/%m/%d')) + ' - ' + str(week_end.strftime('%Y/%m/%d'))
    return df
    
```

```python
def count_avg_plot(df, X, Y, title,  Z=None):
    """
    Generates a plot based of an X and Y and an optional Z axis from a pandas dataframe
    :param df: Pandas Dataframe holding X, Y and Z
    :param X: Pandas Series to be plotted on X axis
    :param Y: Pandas Series to be plotted on Y axis
    :param title: String, title of graph
    :param Z: Pandas Series to be plotted on Z axis
    :return: Matplotlib Plot
    """
    ax = df.plot(x=X, y=Y, legend=False, figsize=(15,10), fontsize = 12, title=title, rot = 45, marker = 'o')
    ax.grid(True)
    ax.set_ylabel(Y)
    #ax.annotate("Point 1", (43, 91249856))
    ax.text(43, 91249856,'Quarantine Begins',horizontalalignment='right', fontsize = 'large')
    ax.title.set_size(20)
    if Z:
        ax2 = ax.twinx()
        ax2.set_ylabel(Z)
        df.plot(x=X, y=Z, ax=ax2, legend=False, color="r", linestyle='--', fontsize=15).grid(False)
        ax2.get_yaxis().set_visible(False)
    ax.figure.legend(loc=4)
    #ax.xticks(fontsize=14)
    plt.show()
```

```python
def get_audio_features(df):
    features_json = spotifyObject.audio_features(tracks = df['song_id'])[0]
    df['danceability'] = features_json['danceability']
    df['energy'] = features_json['energy']
    df['loudness'] = features_json['loudness']
    df['speechiness'] = features_json['speechiness']
    df['acousticness'] = features_json['acousticness']
    df['instrumentalness'] = features_json['instrumentalness']
    df['liveness'] = features_json['liveness']
    df['valence'] = features_json['valence']
    return df
    
```

```python
def playlist_tracks(username, playlist_id):
    playlist_tracks = spotifyObject.user_playlist_tracks(user=username, playlist_id=playlist_id)
    tracks = playlist_tracks['items']
    while spotifyObject.next(playlist_tracks):
        playlist_tracks = spotifyObject.next(playlist_tracks)
        tracks.extend(playlist_tracks['items'])
    return tracks
```

```python
playlist_dict = {}
for val in range(len(playlists['items'])):
    playlist_dict[playlists['items'][val]['name']] = playlist_tracks(username, playlists['items'][val]['id'])
```

```python
def playlist_tracks_df(playlist_d, key):
    df = pd.DataFrame()
    for idx in range(len(playlist_d[key])):
        df.loc[idx, 'Playlist'] = key
        df.loc[idx, 'Added Date'] = playlist_d[key][idx]['added_at']
        df.loc[idx, 'Artist'] = playlist_d[key][idx]['track']['artists'][0]['name']
        df.loc[idx, 'songId'] = playlist_d[key][idx]['track']['id']
        df.loc[idx, 'Song'] = playlist_d[key][idx]['track']['name']
        df.loc[idx, 'Album'] = playlist_d[key][idx]['track']['album']['name']
    return df
        
```
